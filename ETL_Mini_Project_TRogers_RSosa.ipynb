```python
# Create a copy of the crowdfunding_info_df DataFrame name campaign_df. 
crowdfunding_df = pd.read_excel(r"C:\Users\troge\Desktop\Bootcamp\Project 2\Starter_Files\Resources\crowdfunding.xlsx")
campaign_df = crowdfunding_df.copy()
campaign_df.head()
```

```python
# Rename the blurb, launched_at, and deadline columns.
crowdfunding_df.rename(columns={
    'blurb': 'description',
    'launched_at': 'launch_date',
    'deadline': 'end_date'
}, inplace=True)
```

```python
# Convert the goal and pledged columns to a `float` data type.
crowdfunding_df['goal'] = crowdfunding_df['goal'].astype(float)
crowdfunding_df['pledged'] = crowdfunding_df['pledged'].astype(float)
```

```python
# Check the datatypes
crowdfunding_df.dtypes
```

```python
# Format the launched_date and end_date columns to datetime format
from datetime import datetime as dt
crowdfunding_df.rename(columns={'launched_at': 'launch_date', 'deadline': 'end_date'}, inplace=True)
crowdfunding_df['launch_date'] = pd.to_datetime(crowdfunding_df['launch_date'])
crowdfunding_df['end_date'] = pd.to_datetime(crowdfunding_df['end_date'])
```

```python
# Merge the campaign_df with the category_df on the "category" column and 
# the subcategory_df on the "subcategory" column.
crowdfunding_df = pd.DataFrame({'campaign_id': [1, 2, 3],
                                'category': ['Art', 'Technology', 'Film'],
                                'subcategory': ['Painting', 'Gadgets', 'Documentary']})

category_df = pd.DataFrame({'category': ['Art', 'Technology', 'Film'],
                            'category_description': ['Creative work', 'Innovative projects', 'Movie making']})

subcategory_df = pd.DataFrame({'subcategory': ['Painting', 'Gadgets', 'Documentary'],
                               'subcategory_description': ['Creating artwork with paint', 'Electronic gadgets', 'Non-fiction films']})
merged_df = pd.merge(crowdfunding_df, category_df, on='category', how='left')
final_df = pd.merge(merged_df, subcategory_df, on='subcategory', how='left')
print(final_df)
```
